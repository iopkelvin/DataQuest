** Dropping missing values, and comparing % of missing values **

```python
raw_data_count = recent_grads.shape[0] #Amount of rows
recent_grads = recent_grads.dropna() #Dropping missing values
cleaned_data_count = recent_grads.shape[0]
print("Initial amount of rows: ", raw_data_count)
print("Rows withiout missing values: ", cleaned_data_count)
```

* Only 1 row contained missing values 

### Scatter Plot Relations

```python
recent_grads.plot(x="Sample_size", y="Median", kind="scatter", title="Median vs Sample_size")
```

```python
recent_grads.plot(x="Sample_size", y="Unemployment_rate", kind="scatter", title="Unemployment_rate vs Sample_size")
```

```python
recent_grads.plot(x="Full_time", y="Median", kind="scatter", title="Median vs Full_time")
```

```python
recent_grads.plot(x="ShareWomen", y="Unemployment_rate", kind="scatter", title="Unemployment_rate vs ShareWomen")
```

```python
recent_grads.plot(x="Men", y="Median", kind="scatter", title="Median vs Men")
```

```python
recent_grads.plot(x="Women", y="Median", kind="scatter", title="Median vs Women")
```

** Do Students in more popular majors make more money? **
* On Average, students earn a a little below average(median salary) for the most popular majors

** Do students that majored in subjects that were majority female make more money? **
* Unemployment rate for subjects that were majority women (>50%) seems to be equally balanced to those with minority women. There is no apparent correlation

** Is there any link between the number of full-time employees and median salary? **
* Slightly, subjects with higher amount of full time employees seem to be earning higher median salary.

### Histograms

```python
recent_grads["Sample_size"].hist(bins = 15, range=(0,5000))
plt.title("Sample_size Histogram")
plt.xlabel("Sample Size")
plt.ylabel("Distribution")
```

```python
recent_grads["Median"].hist(bins = 30, range=(0,120000))
plt.title("Median Histogram")
plt.xlabel("Median")
plt.ylabel("Distribution")
```

```python
recent_grads["Employed"].hist(bins = 25, range=(0,320000))
plt.title("Employed Histogram")
plt.xlabel("Employed")
plt.ylabel("Distribution")
```

```python
recent_grads["Full_time"].hist(bins = 25, range=(0,260000))
plt.title("Full_time Histogram")
plt.xlabel("Full_time")
plt.ylabel("Distribution")
```

```python
recent_grads["ShareWomen"].hist(bins = 20, range=(0,1))
plt.title("ShareWomen Histogram")
plt.xlabel("ShareWomen")
plt.ylabel("Distribution")
```

```python
recent_grads["Unemployment_rate"].hist(bins = 25, range=(0,0.19))
plt.title("Unemployment_rate Histogram")
plt.xlabel("Unemployment_rate")
plt.ylabel("Distribution")
```

```python
recent_grads["Men"].hist(bins = 15, range=(0,180000))
plt.title("Men Histogram")
plt.xlabel("Men")
plt.ylabel("Distribution")
```

```python
recent_grads["Women"].hist(bins = 15, range=(0,310000))
plt.title("Women Histogram")
plt.xlabel("Women")
plt.ylabel("Distribution")
```

** What percent of majors are predominantly male? Predominantly female? **
* There isn't much difference. An average of 55% jobs are predominantly women. 45% predominantly men

** What's the most common median salary range? **
* The most common median salary range is from 30,000 to 50,000
