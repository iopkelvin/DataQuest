```python
import numpy as np
import pandas as pd

dc_listings = pd.read_csv("dc_airbnb.csv")
stripped_commas = dc_listings['price'].str.replace(',', '')
stripped_dollars = stripped_commas.str.replace('$', '')
dc_listings['price'] = stripped_dollars.astype('float')
```

```python
dc_listings = dc_listings.iloc[np.random.permutation(dc_listings.index)]
```

```python
split_one = dc_listings.iloc[:1862]
split_two = dc_listings.iloc[1862:]
```

## Train/ test KNN 

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error

train_one = split_one
test_one = split_two
train_two = split_two
test_two = split_one
```

```python
knn = KNeighborsRegressor()
knn.fit(train_one[["accommodates"]], train_one[["price"]])
predicted = knn.predict(test_one[["accommodates"]])
mse = mean_squared_error(predicted, test_one["price"])
iteration_one_rmse = mse ** (1/2)
```

```python
knn = KNeighborsRegressor()
knn.fit(train_two[["accommodates"]], train_two[["price"]])
predicted = knn.predict(test_two[["accommodates"]])
mse = mean_squared_error(predicted, test_two["price"])
iteration_two_rmse = mse ** (1/2)
```

```python
avg_rmse = np.mean([iteration_one_rmse, iteration_two_rmse])
```
